```python
pink_data = all_data[['% With Access to Exercise Opportunities', '% Uninsured',
'Primary Care Physicians Rate',
'Dentist Rate',
'Mental Health Provider Rate',
'Preventable Hospitalization Rate',
'% with Annual Mammogram',
'% Vaccinated', '% Adults with Obesity',
'% Adults with Diabetes',
'Life Expectancy',
'Drug Overdose Mortality Rate'
                     ]]
```

```python
pink_columns = ['% With Access to Exercise Opportunities', '% Uninsured',
'Primary Care Physicians Rate',
'Dentist Rate',
'Mental Health Provider Rate',
'Preventable Hospitalization Rate',
'% with Annual Mammogram',
'% Vaccinated', '% Adults with Obesity',
'% Adults with Diabetes',
'Life Expectancy',
'Drug Overdose Mortality Rate'
                     ]
```

```python
target = '% Adults with Diabetes'
```

```python
num_features = 12
num_cols = 3  # Number of columns in the grid
num_rows = (num_features + num_cols - 1) // num_cols  # Calculate the number of rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten() 

# Loop through the features and create scatter plots
for ax, feature in zip(axes, pink_columns):
    ax.scatter(pink_data[target], pink_data[feature])
    ax.set_xlabel(target)
    ax.set_ylabel(feature)
    ax.set_title(f'Scatter Plot of {feature} vs. % Adults with Diabetes')

for i in range(len(pink_columns), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Provided data
data = [0.51, 0.74, 0.55, 1.12, 3.31, 0.46, 2.72, 0.46, 0.43, 2.27, 2.51, 2.34, 0.27, 0.63, 2.28]

# Sort the data in ascending order
sorted_data = sorted(data)

# Create a bar plot to visualize the sorted data
plt.figure(figsize=(10, 6))
sns.barplot(x=np.arange(1, len(sorted_data) + 1), y=sorted_data, palette='Blues_d', edgecolor='black')

# Customize plot
plt.title('Sorted Data Values', fontsize=16)
plt.xlabel('Index (in ascending order)', fontsize=14)
plt.ylabel('Value', fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```
