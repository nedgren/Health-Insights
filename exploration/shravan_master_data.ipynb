```python
dropped_rows = data[data['NaN_Count'] <= 35]
```

```python
dropped_rows = dropped_rows[dropped_rows['County'].notna()]
```

```python
missing_values = data.isnull().sum()

# Filter columns with more than 300 missing values and sort them in descending order
missing_values_filtered = missing_values[missing_values > 300].sort_values(ascending=False) #300 is ~10% of the total number of rows

#print("\nMissing Values per Column (More than 300):")
#print(missing_values_filtered)
```

```python
missing_values_filtered
```

```python
features_to_drop = missing_values_filtered.index.tolist()
```

```python
features_to_drop
```

```python
dropped_both = dropped_rows.drop(columns = features_to_drop)
```

```python
dropped_both
```

```python
dropped_both.info()
```

```python
def impute_with_mean(df):
    for column in df.columns:
        if column != 'State' and column != 'County' and column != 'Presence of Water Violation':  # Skip the 'state' column
            # Calculate the mean of the column grouped by 'state'
            mean_values = df.groupby('State')[column].transform('mean')
            # Fill NaN values with the calculated means
            df[column] = df[column].fillna(mean_values)
    return df

# Impute missing values
dropped_imputed = impute_with_mean(dropped_both)
```

```python
nan_counts = dropped_imputed.isna().sum(axis=1)
dropped_imputed['NaN_Count'] = nan_counts
sorted_data = dropped_imputed.sort_values(by='NaN_Count', ascending=False)
```
