```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#import data
filepath = '../data/data-reduced.csv'
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
data_train, data_test = train_test_split(data,
                                         test_size=0.2,
                                         shuffle=True,
                                         random_state=708)
```

```python
training_set_index = data_train.index
test_set_index = data_test.index
```

```python
#check that target variable is well-distributed in both train and test sets
target = '% Adults with Diabetes'
plt.hist(data_train[target])
plt.hist(data_test[target])

plt.show()
```

```python
data_train.to_csv('../data/data-reduced-train.csv', index=False)
data_test.to_csv('../data/data-reduced-test.csv', index=False)
```
